### ML_V0.2.ipynb

```python
# 데이터 로드

import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor


df = pd.read_csv("merged.csv")

label_cols = ["구", "연도"]
target_cols = "값"

ohe_cols = [ '강남구','강동구','강북구','강서구','관악구','광진구','구로구','금천구',
    '노원구','도봉구','동대문구','동작구','마포구','서대문구','서초구',
    '성동구','성북구','송파구','양천구','영등포구','용산구','은평구',
    '종로구','중구','중랑구'
]

num_features = [
    c for c in df.columns if c not in (label_cols + [target_cols] + ohe_cols)
]
print("초기 숫자형 피처 갯수 : ", len(num_features))
```

### 선형결합 제거 (남자, 여자 컬럼 없애기)

```python
drop_cols = [ "단독주택_남자","단독주택_여자",
    "아파트_남자","아파트_여자",
    "연립주택_남자","연립주택_여자",
    "다세대주택_남자","다세대주택_여자"]

df = df.drop(columns=[c for c in drop_cols if c in df.columns])
num_features = [c for c in num_features if c not in drop_cols]

print("제거후 숫자형 피처 갯수 : " , len(num_features))
```

### VIF 계산

```python
# object 타입 -> float 변환
vif_features = num_features + ["연도"]

for col in vif_features:
    df[col] = pd.to_numeric(df[col], errors='coerce')

X = df[vif_features].astype(float)

vif_list = []
for i, col in enumerate(vif_features):
    vif_score = variance_inflation_factor(X.values, i)
    vif_list.append({"feature": col, "VIF":vif_score})

vif_df = pd.DataFrame(vif_list).sort_values("VIF", ascending=False)
print("1차 VIF 결과>>>")
print(vif_df)
```
